## üîπ SETS ‚Äî COMPLETE FUNCTION COVERAGE

### Creation & Basic Operations

```python
# Creation
s = {1, 2, 3}
s2 = set([3, 4, 5])
empty = set()

print(s, s2, empty)

# add / update
s.add(4)
s.update([5, 6])
print(s)
```

### Removal Operations

```python
# remove vs discard
s.discard(10)   # safe
# s.remove(10)  # KeyError

# pop (removes arbitrary element)
print(s.pop())
print(s)
```

### Copy & Clear

```python
copy_set = s.copy()
s.clear()

print(copy_set)
print(s)
```

### Mathematical Set Operations

```python
a = {1,2,3}
b = {3,4,5}

print("Union:", a.union(b))
print("Intersection:", a.intersection(b))
print("Difference:", a.difference(b))
print("Symmetric Difference:", a.symmetric_difference(b))

# Operators
print(a | b, a & b, a - b, a ^ b)
```

### Relationship Checks

```python
x = {1,2}
y = {1,2,3,4}

print(x.issubset(y))
print(y.issuperset(x))
print(x.isdisjoint({7,8}))
```

### Set Comprehension

```python
nums = [1,2,2,3,4,4,5]
unique_even = {x for x in nums if x % 2 == 0}
print(unique_even)
```

### FrozenSet (Immutable Set)

```python
fs = frozenset([1,2,3])
print(fs)
```

### Advanced Set Implementations

```python
# Duplicate removal
arr = [1,1,2,3,3,4]
print(list(set(arr)))

# Intersection of arrays
a = [1,2,3,4]
b = [3,4,5,6]
print(list(set(a) & set(b)))
```

## üî∏ DICTIONARIES ‚Äî COMPLETE FUNCTION COVERAGE

### Creation & Access

```python
d = {"a": 1, "b": 2}
print(d["a"])
print(d.get("c", 0))
```

### keys(), values(), items()

```python
print(list(d.keys()))
print(list(d.values()))
print(list(d.items()))
```

### update(), pop(), popitem(), clear()

```python
d.update({"c": 3})
print(d.pop("b"))

# popitem ‚Üí LIFO (Python 3.7+)
print(d.popitem())

d.clear()
print(d)
```

### setdefault() & fromkeys()

```python
d = {}
print(d.setdefault("x", 10))
print(d)

keys = ["a", "b", "c"]
new_dict = dict.fromkeys(keys, 0)
print(new_dict)
```

### Dictionary Comprehension

```python
nums = [1,2,3,4]
squares = {x: x*x for x in nums}
even_squares = {x: x*x for x in nums if x % 2 == 0}
print(squares)
print(even_squares)
```

### Sorting Dictionaries

```python
data = {"apple": 3, "banana": 1, "cherry": 2}

# sort by key
print(dict(sorted(data.items())))

# sort by value
print(dict(sorted(data.items(), key=lambda x: x[1])))
```

### Advanced collections

```python
from collections import defaultdict, Counter, OrderedDict

# defaultdict
dd = defaultdict(int)
for c in "mississippi":
    dd[c] += 1
print(dd)

# Counter
cnt = Counter("mississippi")
print(cnt)
print(cnt.most_common(2))

# OrderedDict
od = OrderedDict()
od["x"] = 1
od["y"] = 2
print(od)
```

### Dictionary as Frequency Map

```python
arr = [1,2,2,3,3,3]
freq = {}

for x in arr:
    freq[x] = freq.get(x, 0) + 1

print(freq)
```

### Inverting a Dictionary

```python
d = {"a": 1, "b": 2, "c": 1}
inv = {}

for k, v in d.items():
    inv.setdefault(v, []).append(k)

print(inv)
```

### Hash Map Algorithms

```python
# Two Sum
def two_sum(nums, target):
    seen = {}
    for i, n in enumerate(nums):
        diff = target - n
        if diff in seen:
            return [seen[diff], i]
        seen[n] = i

print(two_sum([2,7,11,15], 9))

# Group Anagrams
from collections import defaultdict

def group_anagrams(words):
    groups = defaultdict(list)
    for w in words:
        key = tuple(sorted(w))
        groups[key].append(w)
    return list(groups.values())

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```
