### Stack & Queue Behavior

### Prefix Sum

```python
arr = [1, 2,45, 3, 4]
prefix = [0]

for x in arr:
    prefix.append(prefix[-1] + x)
print(prefix)
l, r = 0, 2
print(prefix[r+1] - prefix[l])
```

### Sliding Window

```python
arr = [2,1,5,1,3,2]
k = 3

window_sum = sum(arr[:k])
max_sum = window_sum

for i in range(k, len(arr)):
    window_sum += arr[i] - arr[i-k]
    max_sum = max(max_sum, window_sum)

print(max_sum)
```

### Two Pointers

```python
arr = [1,2,3,4,5]
l, r = 0, len(arr)-1

while l < r:
    print(arr[l], arr[r])
    l += 1
    r -= 1
```

## 8Ô∏è‚É£ ADVANCED ALGORITHMS

```python
# Balanced Parentheses
def isValid(s):
    stack = []
    mp = {')':'(', ']':'[', '}':'{'}

    for c in s:
        if c in mp:
            if not stack or stack.pop() != mp[c]:
                return False
        else:
            stack.append(c)

    return not stack

print(isValid("()[]{}"))
```

```python
# Anagram Check
from collections import Counter

def isAnagram(s, t):
    return Counter(s) == Counter(t)

print(isAnagram("listen", "silent"))
```
