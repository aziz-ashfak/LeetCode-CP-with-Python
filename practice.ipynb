```python
def isAnagram( s, t) :
        if len(s) != len(t):
            return False
        
        count = {}
        
        for ch in s:
            count[ch] = count.get(ch, 0) + 1
        
        for ch in t:
            if ch not in count:
                return False
            count[ch] -= 1
            if count[ch] == 0:
                del count[ch]
        
        return len(count) == 0
```

```python
s = "anagram"
t = "nagaram"
isAnagram(s,t)
```

```python
s="eat"
t="tea"
isAnagram(s,t)
```

```python
def anagram(a, b):
    print("Input strings:")
    print("a =", a)
    print("b =", b)
    print("-" * 30)

    # Step 1: Length check
    print("Checking lengths...")
    if len(a) != len(b):
        print("Lengths are different → Not anagrams")
        return False
    print("Lengths are equal ✓")
    print("-" * 30)

    # Step 2: Initialize dictionary
    count = {}
    print("Initial count dictionary:", count)
    print("-" * 30)

    # Step 3: Count characters in 'a'
    print("Counting characters from string a:")
    for ch in a:
        count[ch] = count.get(ch, 0) + 1
        print(f"Added '{ch}' → {count}")
    print("-" * 30)

    # Step 4: Subtract characters using 'b'
    print("Subtracting characters using string b:")
    for ch in b:
        if ch not in count:
            print(f"'{ch}' not found → Not anagrams")
            return False

        count[ch] -= 1
        print(f"Subtracted '{ch}' → {count}")

        if count[ch] < 0:
            print(f"Count of '{ch}' became negative → Not anagrams")
            return False
    print("-" * 30)

    # Step 5: Final result
    print("All counts matched → Anagrams ✓")
    return True


# Run the function
print("Final Result:", anagram("cat", "tac"))
```

```python
def anagram_26(a, b):
    print("Input strings:")
    print("a =", a)
    print("b =", b)
    print("-" * 50)

    # Step 1: Length check
    print("Checking lengths...")
    if len(a) != len(b):
        print("❌ Length mismatch → Not anagrams")
        return False
    print("✅ Length matched")
    print("-" * 50)

    # Step 2: Create 26-length array
    count = [0] * 26
    print("Initial 26-char array:")
    print(count)
    print("-" * 50)

    # Step 3: Count characters from 'a'
    print("Adding characters from string a:")
    for ch in a:
        idx = ord(ch) - ord('a')
        count[idx] += 1
        print(f"Added '{ch}' at index {idx} → {count}")
    print("-" * 50)

    # Step 4: Subtract characters using 'b'
    print("Subtracting characters from string b:")
    for ch in b:
        idx = ord(ch) - ord('a')
        count[idx] -= 1
        print(f"Subtracted '{ch}' at index {idx} → {count}")

        if count[idx] < 0:
            print(f"❌ Count went negative for '{ch}' → Not anagrams")
            return False
    print("-" * 50)

    # Step 5: Final check
    print("Final 26-char array:")
    print(count)
    print("✅ All zeros → Anagrams")
    return True


# Run example
print("Final Result:", anagram_26("cat", "tac"))
```
