# 🐍 Python Strings — Complete join / split / text utility Functions

This notebook section focuses **only on string-related functions**:
- join, split, rsplit, splitlines
- partition, rpartition
- replace, translate
- strip, lstrip, rstrip
- find, index, rfind, rindex
- count
- startswith, endswith
- zfill, center, ljust, rjust
- ord, chr
- os.path string helpers (splitext, basename, dirname)

Perfect as a **string master cheat sheet**.


## 🔗 join()

```python
words = ["Python", "is", "awesome"]

print(" ".join(words))
print("-".join(words))
print("".join(words))
```

## ✂ split(), rsplit(), splitlines()

```python
s = "one,two,three,four"

print(s.split(","))
print(s.split(",", 1))
print(s.rsplit(",", 1))

multi = "hello\nworld\npython"
print(multi.splitlines())
```

## 🧩 partition() / rpartition()

```python
text = "user@example.com"

print(text.partition("@"))
print(text.rpartition("."))
```

## 🧼 strip(), lstrip(), rstrip()

```python
s = "***hello***"

print(s.strip("*"))
print(s.lstrip("*"))
print(s.rstrip("*"))
```

## 🔍 find(), index(), rfind(), rindex()

```python
s = "banana"

print(s.find("na"))
print(s.rfind("na"))

print(s.index("na"))
print(s.count("a"))
```

## 🚦 startswith(), endswith()

```python
filename = "data.csv"

print(filename.startswith("data"))
print(filename.endswith(".csv"))
```

## 📐 justify & zfill

```python
s = "42"

print(s.zfill(5))
print("hi".center(10, "*"))
print("hi".ljust(10, "-"))
print("hi".rjust(10, "-"))
```

## 🔁 replace() & translate()

```python
s = "hello world"
print(s.replace("world", "python"))

table = str.maketrans("aeiou", "12345")
print("education".translate(table))
```

## 🔢 ord() / chr()

```python
print(ord("a"))
print(chr(97))
```

## 📁 Path-related string helpers (os.path)

```python
import os

path = "/home/user/data/file.txt"

print(os.path.basename(path))
print(os.path.dirname(path))
print(os.path.splitext(path))
```
